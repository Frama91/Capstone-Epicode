```python
# Importo le librerie necessari per svolgere le attività

import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np


# Faccio in modo che Pandas consideri le celle vuote come valori "NaN"
pd.options.mode.use_inf_as_na = True

Folder_Path = "C:/Users/franc/Desktop/Portfolio/Python/"
```

# Data cleaning con Python

Utilizzando la libreria pandas, effettueremo un controllo sui datatbase ed una successiva pulizia dei dati in modo da renderli usufruibili per le analisi
I passaggio sono i seguenti:

**-Eliminazione colonne con dati non necessari**

**-Eliminazione righe con dati mancanti**

**-Controllo duplicati**

**-Controllo finale**


```python
# DB Employees: Carichiamo il file CSV tramite il metodo read_csv() di pandas e diamo una prima occhiata ai dati 
DB_Employees = pd.read_csv("C:/Users/franc/Desktop/Portfolio/Origine Dati/DF_Dipendenti.csv")
display(DB_Employees.sample(10))

# Ai fini di una futura analisi, che faccia il punto sul rendimento del dipendente, possiamo eliminare le colonne "Telephone" ed "Email"
Clean_Employees = DB_Employees.drop(columns=["Telephone","Email"])

# A questo punto eliminiamo le righe con dati mancanti e successivamente effettuiamo il controllo
Visual = pd.DataFrame(columns= [Clean_Employees.columns]).T
Visual["Pre Clean Null Values"] = list(Clean_Employees.isnull().sum())
old_row = Clean_Employees.shape[0]
Clean_Employees.dropna(inplace=True)
Visual["After Clean Null Values"] = list(Clean_Employees.isnull().sum())
new_row = Clean_Employees.shape[0]
display(Visual)
print("Il numero di righe prima della pulizia era",old_row,"\nAdesso il totale è: ",new_row)

# Ripristiniamo l'index nel database
Clean_Employees.reset_index(inplace=True,drop="index")

# Per scrupolo controlliamo anche la presenza nella colonna "Code", che sappiamo dovrebbe contenere solo valori univoci
print("Il numero di valori unici, nella colonna 'Code', è",Clean_Employees["Code"].nunique(),"\nEssendo il numero di valori unici uguale al numero di righe totali, deduciamo che non ci sono duplicati.")

# Risultato Finale
display(Clean_Employees.sample(10))

# Salvo i dati puliti in un nuovo file CSV
Clean_Employees.to_csv(Folder_Path+"Clean_Employees.csv",index=False)
```

```python
# DB Products: Carichiamo il file CSV tramite il metodo read_csv() di pandas e diamo una prima occhiata ai dati 
DB_Products = pd.read_csv("C:/Users/franc/Desktop/Portfolio/Origine Dati/DF_Prodotti.csv")
display(DB_Products.sample(10))

# In questo caso manteniamo tutte le colonne perchè possono risultarte utili ai fini del report

# Controlliamo ed eliminiamo i valori nulli con un metodo differente.
# In questo caso non ci vengono conteggiati i valori nulli ma ci viene detto se ci sono o meno
Clean_Products = DB_Products
old_row = len(Clean_Products["ID Product"])
Visual = pd.DataFrame(columns= [Clean_Products.columns]).T
Visual["Pre Clean Null Values"] = list(Clean_Products.isnull().any())
Clean_Products.dropna(inplace=True)
Visual["After Clean Null Values"] = list(Clean_Products.isnull().any())
new_row = len(Clean_Products)


display(Visual)
print("Il numero di righe prima della pulizia era",old_row,"\nAdesso il totale è: ",new_row)

# Ripristiniamo l'index nel database
Clean_Products.reset_index(inplace=True,drop="index")

# Controlliamo i duplicati su "ID Product"
print("Il numero di valori dublicati sulla colonna 'ID Product' è",Clean_Products["ID Product"].duplicated().sum())

# Risultato Finale
display(Clean_Products.sample(10))

# Salvo i dati puliti in un nuovo file CSV
Clean_Products.to_csv(Folder_Path+"Clean_Products.csv",index=False)
```

```python
# DB Clients: Carichiamo il file CSV tramite il metodo read_csv() di pandas e diamo una prima occhiata ai dati 
DB_Clients = pd.read_csv("C:/Users/franc/Desktop/Portfolio/Origine Dati/DF_Clienti.csv")
display(DB_Clients)

# Controllo dei valori nulli e confronto
Clean_Clients = DB_Clients.copy()
old_row = len(Clean_Clients["ID Client"])
Visual_Clients = pd.DataFrame(columns=[Clean_Clients.columns]).T
Visual_Clients["Pre Clean Null Values"] = list(Clean_Clients.isna().sum())
Clean_Clients.dropna(inplace=True)
Visual_Clients["After Clean Null Values"] = list(Clean_Clients.isna().sum())
new_row = len(Clean_Clients)

display(Visual_Clients)
print("Il numero di righe prima della pulizia era", old_row, "\nAdesso il totale è: ", new_row)

# Ripristiniamo l'index nel database
Clean_Clients.reset_index(inplace=True, drop="index")

# Controlliamo i duplicati su "ID Client"
print("Il numero di valori duplicati sulla colonna 'ID Client' è", Clean_Clients["ID Client"].duplicated().sum())

# Risultato Finale
display(Clean_Clients)

# Salvo i dati puliti in un nuovo file CSV
Clean_Clients.to_csv(Folder_Path+"Clean_Clients.csv",index=False)
```

```python
# DB Sales: Carichiamo il file CSV tramite il metodo read_csv() di pandas e diamo una prima occhiata ai dati 
DB_Sales = pd.read_csv("C:/Users/franc/Desktop/Portfolio/Origine Dati/DF_Vendite.csv")
display(DB_Sales.sample(10))

# Controlliamo ed eliminiamo i valori nulli con un metodo differente.
# In questo caso non ci vengono conteggiati i valori nulli ma ci viene detto se ci sono o meno
Clean_Sales = DB_Sales.copy()
old_row = len(Clean_Sales["ID Sale"])
Visual_Sales = pd.DataFrame(columns=[Clean_Sales.columns]).T
Visual_Sales["Pre Clean Null Values"] = list(Clean_Sales.isnull().sum())
Clean_Sales.dropna(inplace=True)
Visual_Sales["After Clean Null Values"] = list(Clean_Sales.isnull().sum())
new_row = len(Clean_Sales)

display(Visual_Sales)
print("Il numero di righe prima della pulizia era", old_row, "\nAdesso il totale è: ", new_row)

# Ripristiniamo l'index nel database
Clean_Sales.reset_index(inplace=True, drop="index")

# Controlliamo i duplicati su "ID Sale"
print("Il numero di valori duplicati sulla colonna 'ID Sale' è", Clean_Sales["ID Sale"].duplicated().sum())

# Risultato Finale
display(Clean_Sales.sample(10))

# Salvo i dati puliti in un nuovo file CSV
Clean_Sales.to_csv(Folder_Path+"Clean_Sales.csv",index=False)
```

## Visualizzazione dati

Creiamo degli esempi basici di visualizzazione dei dati su dei grafici

```python
# Grafico a barre orizzontali sui dati sugli stipendi in base al ruolo ed esperienza

# Calcolo della media per Ruolo ed Esperienza
Salary_Mean = Clean_Employees.groupby(["Role", "Experience"])["Salary USD"].mean()
Salary_Mean = round(Salary_Mean, 2)

# Creazione del Database
Salary_Mean = Salary_Mean.reset_index()

# Definizione dell'ordine personalizzato per l'esperienza
Exp_ord = ['Junior', 'Intermediate', 'Senior', 'Lead/Manager']

# Conversione della colonna 'Experience' in variabile categorica con ordine specifico
Salary_Mean['Experience'] = pd.Categorical(Salary_Mean['Experience'], categories=Exp_ord, ordered=True)

# Ordinamento del DataFrame per "Role" e "Experience"
Salary_Mean = Salary_Mean.sort_values(by=["Role", "Experience"])

# Creazione del grafico a barre orizzontali con gruppi di colonne
fig, ax = plt.subplots(figsize=(20,15))

# Creazione di un indice numerico per ogni Role
role_index = np.arange(len(Salary_Mean['Role'].unique()))

# Larghezza delle colonne
bar_width = 0.2

# Iterazione sui livelli di esperienza univoci
for i, exp in enumerate(Salary_Mean['Experience'].unique()):
    # Calcolo della posizione delle colonne per ciascun livello di esperienza
    positions = role_index + i * bar_width

    # Estrazione dei dati per l'esperienza corrente
    exp_data = Salary_Mean[Salary_Mean['Experience'] == exp]

    # Creazione delle colonne per l'esperienza corrente
    ax.barh(positions, exp_data['Salary USD'], height=bar_width, label=exp)

# Personalizzazione dell'aspetto del grafico
ax.set_yticks(role_index + (len(Salary_Mean['Experience'].unique()) - 1) * bar_width / 2)
ax.set_yticklabels(Salary_Mean['Role'].unique())
ax.set_xlabel('Salary USD')
ax.set_ylabel('Role')
ax.set_title('Salaries by Role and Experience')
ax.legend(title='Experience', bbox_to_anchor=(1.05, 1), loc='upper left')

# Aggiunta di una griglia sull'asse x con stile tratteggiato e trasparenza
ax.grid(axis='x', linestyle='-', alpha=1)

# Personalizzazione dei colori delle barre
colors = ['blue', 'green', 'orange', 'red']
for i, exp in enumerate(Salary_Mean['Experience'].unique()):
    positions = role_index + i * bar_width
    exp_data = Salary_Mean[Salary_Mean['Experience'] == exp]
    ax.barh(positions, exp_data['Salary USD'], height=bar_width, label=exp, color=colors[i])


# Salva e visualizza il grafico
plt.savefig(Folder_Path+"Barre_Stipendi.jpeg")
plt.show()
```

```python
# Serie storica sul totale delle quantità vendute, grafico a linea.

# Converti 'Sale Date' in formato data
Clean_Sales['Sale Date'] = pd.to_datetime(Clean_Sales['Sale Date'], format='%d/%m/%Y')

# Raggruppa per mese e calcola il totale delle quantità vendute
total_quantity_sold_monthly = Clean_Sales.resample('M', on='Sale Date')['Quantity Sold'].sum()

# Creazione del grafico a linea
plt.figure(figsize=(20, 10))
plt.plot(total_quantity_sold_monthly.index, total_quantity_sold_monthly.values, marker='o', linestyle='-', color='#2980e3')

# Personalizzazione del grafico
plt.title('Serie Storica delle Quantità Vendute (Mensile)')
plt.xlabel('Data di Vendita')
plt.ylabel('Quantità Vendute')
plt.grid(True)

# Salva e visualizza il grafico
plt.savefig(Folder_Path + "Serie_temporale_mensile.jpeg")
plt.show()
```
