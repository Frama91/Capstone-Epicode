## Merge delle tabelle

#### Creo un unico file con tutti i dati che userò per creare i report.
#### Essendo la quantità di dati abbastanza contenuta, ciò mi permette di gestire meglio il tutto.

```python

import pandas as pd

clients = pd.read_csv("C:/Users/franc/Desktop/Portfolio/Python/Clean_Clients.csv")
employees = pd.read_csv("C:/Users/franc/Desktop/Portfolio/Python/Clean_Employees.csv")
products = pd.read_csv("C:/Users/franc/Desktop/Portfolio/Python/Clean_Products.csv")
sales =  pd.read_csv("C:/Users/franc/Desktop/Portfolio/Python/Clean_Sales.csv")
```

```python
# Comincio creando un nuovo dataset che conterrà la prima unione per poi aggiornarlo con le unioni successive
total_data = pd.merge(sales, clients,left_on="ID Client", right_on="ID Client")
total_data = pd.merge(total_data,employees, left_on="Seller Code", right_on="Code")
total_data = pd.merge(total_data,products,left_on="ID Product",right_on="ID Product")

# Elimino le colonne superflue o duplicate
total_data = total_data.drop(columns=["Code","Role"])

```

```python
# Creo il file CSV
total_data.to_csv("C:/Users/franc/Desktop/Portfolio/Looker-Power BI/Total_data.csv",index = False)
```
